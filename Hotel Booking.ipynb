## Data Analysis and Visualizations

```python
#Calculating the percentage of values in the 'is_canceled' column

canceled_percentage = df['is_canceled'].value_counts(normalize = True)
canceled_percentage
```

```python
#Displaying a bar plot showing the count of reservation status

plt.figure(figsize = (4,4))
plt.title("Reservation status count", size = 13)
plt.xlabel("Cancelations")
plt.ylabel("Number of reservations")
plt.bar(['Not canceled', 'Canceled'],df['is_canceled'].value_counts(), color = 'c', edgecolor = 'k', width = 0.3)
#plt.legend()
plt.show()
```

```python
plt.figure(figsize = (4,4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Greens')
#legend_labels,_ = ax1.get_legend_handles_labels()
#print(legend_labels)
ax1.legend(bbox_to_anchor = (0,1))
plt.title("Reservation status in different hotels", size = 13)
plt.xlabel("hotel")
plt.ylabel("Number of reservations")
plt.legend(["Not canceled","Canceled"])
plt.show()
```

```python
# Calculating the percentage of values for reservations made at the 'Resort Hotel'

resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
# Calculating the percentage of values for reservations made at the 'City Hotel'

city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
# Grouping data by reservation status date and calculating the mean of average daily rates for Resort Hotel and City Hotel

resort_hotel = resort_hotel.groupby('reservation_status_date')[['average_daily_rate']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['average_daily_rate']].mean()
```

```python
# Plotting average daily rates for Resort Hotel and City Hotel

plt.figure(figsize = (25,8))
plt.title("Average daily rate in Resort and City hotel", size = 30)
plt.xlabel("Reservation status date", fontsize = 20)
plt.ylabel("Average daily rate", fontsize = 20)
plt.plot(resort_hotel.index, resort_hotel["average_daily_rate"], label = "Resort Hotel")
plt.plot(city_hotel.index, city_hotel["average_daily_rate"], label = "City Hotel")
plt.legend(fontsize = 25)
plt.show()
```

```python
#Checking the maximum reservation and cancellations in a month

df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = df, palette = 'Set2')

# Creating a dictionary mapping month numbers to month names
month_names = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}

# Changing the month labels on the x-axis
ax1.set_xticklabels([month_names[x] for x in range(1, 13)])

ax1.legend(bbox_to_anchor = (1,1))
plt.title('Reservations status per month', size = 20)
plt.xlabel('Month', fontsize = 14)
plt.ylabel('Number of reservations', fontsize = 14)
plt.legend(['Not Canceled', 'Canceled'])
plt.show()
```

```python
#Plotting average daily rate for each month

plt.figure(figsize = (15, 8))
plt.title("Average daily rate per month", size = 20)
sns.barplot(x = 'month', y = 'average_daily_rate', data = df[df['is_canceled'] == 1].groupby('month')[['average_daily_rate',]].sum().reset_index())
plt.xlabel("Month", fontsize = 12)
plt.ylabel("Average daily rate", fontsize = 12)
plt.show()
```

```python
canceled_data = df[df["is_canceled"] == 1]
top_10_countries = canceled_data['country'].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.title('Top 10 countries with canceled reservations')
plt.pie(top_10_countries, autopct = '%.2f%%', labels = top_10_countries.index)
plt.show()
```

```python
#Checking from where majority of clients are coming from

df['market_segment'].value_counts()
```

```python
#Checking the percentage of majority of the clints are coming from

df['market_segment'].value_counts(normalize = True)
```

```python
#Checking the percentage of canceled bookings of both hotels

canceled_data['market_segment'].value_counts(normalize = True)
```

```python
#Checking the cancellations of both hotels

canceled_df_adr = canceled_data.groupby('reservation_status_date')[['average_daily_rate']].mean()
canceled_df_adr.reset_index(inplace = True)
canceled_df_adr.sort_values('reservation_status_date', inplace = True)

not_canceled_data = df[df["is_canceled"] == 0]
not_canceled_df_adr = not_canceled_data.groupby('reservation_status_date')[['average_daily_rate']].mean()
not_canceled_df_adr.reset_index(inplace = True)
not_canceled_df_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize = (20,6))
plt.title("Average Daily Rate", size = 22)
plt.xlabel("Reservation status date", fontsize = 16)
plt.ylabel("Average daily rate", fontsize = 16)
plt.plot(not_canceled_df_adr["reservation_status_date"],not_canceled_df_adr['average_daily_rate'], label = 'not canceled')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['average_daily_rate'], label = 'canceled')
plt.legend(fontsize = 15)
```

```python
# Filter the DataFrame to include only data within the specified date range (from '2016' to '2017-09') for canceled and not canceled reservations

canceled_df_adr = canceled_df_adr[(canceled_df_adr['reservation_status_date'] > '2016') & (canceled_df_adr['reservation_status_date'] < '2017-09')]
not_canceled_df_adr = not_canceled_df_adr[(not_canceled_df_adr['reservation_status_date'] > '2016') & (not_canceled_df_adr['reservation_status_date'] < '2017-09')]
```

```python
plt.figure(figsize = (20,6))
plt.title("Average Daily Rate", size = 25)
plt.xlabel("Reservation status date", fontsize = 16)
plt.ylabel("Average daily rate", fontsize = 16)
plt.plot(not_canceled_df_adr["reservation_status_date"],not_canceled_df_adr['average_daily_rate'], label = 'not canceled')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['average_daily_rate'], label = 'canceled')
plt.legend(fontsize = 15)
plt.show()
```
